### Converting categorical data into numerical data

```python
df['Species'].replace({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3}, inplace=True)
df
```

### Splitting data into training and test dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']], df['Species'], test_size=0.2)
```

```python
print("Xtrain lenght: ", len(x_train))
print("Xtest lenght: ", len(x_test))
print("ytrain lenght: ", len(y_train))
print("ytest lenght: ", len(y_test))
```

```python

print("Xtrain shape: ",x_train.shape)
print("Xtest shape: ", x_test.shape)
print("ytrain shape: ",y_train.shape)
print("ytest shape: ", y_test.shape)
```

### Building Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

# Creating object of Logistic Regression
lr = LogisticRegression()
```

###  Training dataset

```python
lr.fit(x_train, y_train)
```

### Predicting values

```python
lr.predict(x_test)

lr.predict([[8.1,4.8,2.5,1.3]])

y_pred = lr.predict(x_test)
```

### Checking accuracy

```python
lr.score(x_test, y_test)
```

```python
from sklearn.metrics import accuracy_score

acc2 = accuracy_score(y_test, y_pred)
acc2
```

```python
acc3 = accuracy_score(y_test, y_pred)
acc3
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
